```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = Path("Resources/athletic_sales_2020.csv")
athletic_sales_2021 = Path("Resources/athletic_sales_2021.csv")

athletic_sales_2020_df = pd.read_csv(athletic_sales_2020)
athletic_sales_2021_df = pd.read_csv(athletic_sales_2021)
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows").reset_index(drop=True)
combined_sales_df
```

```python
# Check if any values are null.
combined_sales_df.isnull().sum()
```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df["invoice_date"] = pd.to_datetime(combined_sales_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# create new dataframe and exclude invoice_date column
products_sold = combined_sales_df.groupby(["region", "state", "city"]).sum(numeric_only=True).rename(columns={"quantity": "Total_Products_Sold"})

# Show the top 5 results.
products_sold
```

#### Using `pivot_table`

```python
# using df.pivot_table() to pivot the data
# Show the number products sold for region, state, and city.
combined_sales_df.pivot_table(index=["region", "state", "city"], values="units_sold", aggfunc="sum")

# Rename the "units_sold" column to "Total_Products_Sold"
renamed_df = combined_sales_df.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
renamed_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# using groupby() to group the data
# Show the total sales for the products sold for each region, state, and city.
# create new dataframe and exclude invoice_date column
total_sales = combined_sales_df.groupby(["region", "state", "city", "invoice_date"]).sum(numeric_only=True)

# Rename the "total_sales" column to "Total Sales"
total_sales = total_sales.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# using pivot_table() to pivot the data
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# create new dataframe and exclude invoice_date column
total_sales_pivot = combined_sales_df.pivot_table(index=["region", "state", "city"], values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot = total_sales_pivot.rename(columns={"total_sales": "Total Sales"})


# Show the top 5 results.
total_sales_pivot.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# using groupby() to group the data
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailer = combined_sales_df.groupby(["retailer", "region", "state", "city"]).sum(numeric_only=True)

# Rename the "total_sales" column to "Total Sales"
total_sales_retailer = total_sales_retailer.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
total_sales_retailer.head()
```

#### Using `pivot_table`

```python
# using pivot_table() to pivot the data
# Show the total sales for the products sold for each retailer, region, state, and city.
# create new dataframe and exclude invoice_date column
total_sales_retailer_pivot = combined_sales_df.pivot_table(index=["retailer", "region", "state", "city"], values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_retailer_pivot = total_sales_retailer_pivot.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
total_sales_retailer_pivot.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = combined_sales_df[combined_sales_df["product"].str.contains("Women's Athletic Footwear", case=False, na=False)]
womens_footwear_sales
```

#### Using `groupby`

```python
# using groupby() to group the data
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
retailer_sales = womens_footwear_sales.groupby(["retailer", "region", "state", "city"]).sum(numeric_only=True)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
retailer_sales = retailer_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
retailer_sales.head()
```

#### Using `pivot_table`

```python
# using pivot_table() to pivot the data
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
retailer_sales_pivot = womens_footwear_sales.pivot_table(index=["retailer", "region", "state", "city"], values="units_sold", aggfunc="sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
retailer_sales_pivot = retailer_sales_pivot.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
retailer_sales_pivot.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
date_sales_pivot = womens_footwear_sales.pivot_table(index="invoice_date", values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
date_sales_pivot = date_sales_pivot.rename(columns={"total_sales": "Total Sales"})

# Show the table.
date_sales_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = date_sales_pivot.resample("D").sum()
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales = daily_sales.sort_values(by="Total Sales", ascending=False)
daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = date_sales_pivot.resample("W").sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales = weekly_sales.sort_values(by="Total Sales", ascending=False)
weekly_sales
```
